```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import re
import pandas as pd
```

```python
driver = webdriver.Chrome()
jobs = []
# numJobs = 100
```

```python
# funtion loading all jobs and passing sign up pop up window
def loadAllJobs():
    while(True):
        try:
            # find the joblist button and click to load jobs
            more = driver.find_element(by = By.CSS_SELECTOR, value = '.JobsList_buttonWrapper__ticwb > button:nth-child(1)')
            more.click()
            time.sleep(2) # sleep time b4 the next action
            try:
                close = driver.find_element(by = By.CSS_SELECTOR, value = '.CloseButton')
                close.click()
                print('sign up pop up closed') # for debugging
            except: 
                print('no pop up...loading jobs')
        except:
            print('all jobs loaded')
            break
```

```python
# function closing cookies if exist
def closeCookies():
    try:
        cookies = driver.find_element(by = By.XPATH, value = '//*[@id="onetrust-accept-btn-handler"]')
        cookies.click()
    except:
        print('no cookies button found')
```

```python
# take job data
def getJobs(url):
    print(f'looking for jobs in {url}') # tell us the code running, {url} is for the current url in urls
    
    driver.get(url)
    
    time.sleep(2)
    
    closeCookies()
    
    loadAllJobs()
    
    job_listings = driver.find_elements(by = By.CSS_SELECTOR, value = 'ul.JobsList_jobsList__lqjTr > li')
    
    print(f'number of jobs: {len(job_listings)}')
    
    for i in range(0, len(job_listings)):
        try:
            job_listings[i].click()
            time.sleep(1)
            deets_xpath = '/html/body/div[3]/div[2]/div[4]/div[2]/div[2]/div/div[1]/header/div[1]'
            
            deets = driver.find_element(by = By.XPATH, value = deets_xpath).text
            
            lines = deets.split('\n')

            employer = lines[0]
            
            try:
                pattern = r'\b\d+\.\d+\b'
                rating = re.findall(pattern, deets)
                title = lines[2]
                location = lines[3]
            except:
                rating = 'null'
                title = lines[1]
                location = lines[2]

            try:
                salary = driver.find_element(by = By.CSS_SELECTOR, 
                                             value = '.SalaryEstimate_medianEstimate__fOYN1').text
            except:
                salary = 'null'

            try:
                size = driver.find_element(by = By.CSS_SELECTOR, 
                                           value = 'div.JobDetails_overviewItem__cAsry:nth-child(1) > div:nth-child(2)').text
            except:
                size = 'null'

            try:
                founded = driver.find_element(by = By.CSS_SELECTOR, 
                                              value = 'div.JobDetails_overviewItem__cAsry:nth-child(2) > div:nth-child(2)').text
            except:
                founded = 'null'

            try:
                type = driver.find_element(by = By.CSS_SELECTOR, 
                                           value = 'div.JobDetails_overviewItem__cAsry:nth-child(3) > div:nth-child(2)').text
            except:
                type = 'null'

            try:
                industry = driver.find_element(by = By.CSS_SELECTOR, 
                                               value = 'div.JobDetails_overviewItem__cAsry:nth-child(4) > div:nth-child(2)').text
            except:
                industry = 'null'

            try:
                sector = driver.find_element(by = By.CSS_SELECTOR, 
                                             value = 'div.JobDetails_overviewItem__cAsry:nth-child(5) > div:nth-child(2)').text
            except:
                sector = 'null'

            try:
                revenue = driver.find_element(by = By.CSS_SELECTOR, 
                                              value = 'div.JobDetails_overviewItem__cAsry:nth-child(6) > div:nth-child(2)').text
            except:
                revenue = 'null'

            try:
                expand_desc = driver.find_element(by = By.CSS_SELECTOR, value = '.JobDetails_showMore___Le6L')
                expand_desc.click()
                time.sleep(3)
                desc_xpath = '/html/body/div[3]/div[2]/div[4]/div[2]/div[2]/div/div[1]/section/div[2]/div[1]'
                job_desc = driver.find_element(by = By.XPATH, value = desc_xpath).text
            except:
                job_desc = 'null' 

            try:
#                 xpath = f'/html/body/div[3]/div[2]/div[4]/div[2]/div[1]/div[2]/ul/li[{i+1}]/div/div/div[1]/div[1]/div[4]/div[2]'
                skills = driver.find_element(by = By.CSS_SELECTOR, 
                                             value = 'div.JobCard_jobDescriptionSnippet__l1tnl > div:nth-child(2)').text
            except:
                skills = 'null'

            jobs.append({"Job Title": title,
                "Company Name": employer,
                "Salary Estimate": salary,
                "Rating": rating,
                "Job Description": job_desc,
                "Skills": skills,
                "Location": location,
                "Founded": founded,
                "Industry": industry,
                "Sector": sector,
                "Revenue": revenue,
                "Size": size,
                "Ownership Type": type})

            print(f'Success - {len(jobs)}')

        except:
            try:
                close = driver.find_element(by = By.CSS_SELECTOR, value = '.CloseButton')
                close.click()
            except:
                deets = 'NA'
                print('closing')
```

```python
# url = 'https://www.glassdoor.com.au/Job/australia-data-science-jobs-SRCH_IL.0,9_IN16_KO10,22.htm'
dataScience = 'https://www.glassdoor.com.au/Job/jobs.htm?sc.occupationParam=%22data+science%22&sc.locationSeoString=Australia&locId=16&locT=N'
dataEngineer = 'https://www.glassdoor.com.au/Job/jobs.htm?sc.occupationParam=%22data+engineer%22&sc.locationSeoString=Australia&locId=16&locT=N'
dataScientist = 'https://www.glassdoor.com.au/Job/jobs.htm?sc.occupationParam=+%22data+scientist%22&sc.locationSeoString=Australia&locId=16&locT=N'
dataAnalyst = 'https://www.glassdoor.com.au/Job/jobs.htm?sc.occupationParam=%22data+analyst%22&sc.locationSeoString=Australia&locId=16&locT=N'

urls = [dataScience, dataEngineer, dataScientist, dataAnalyst]

for url in urls:
    getJobs(url)

driver.quit()
```
